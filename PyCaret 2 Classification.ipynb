# 1. Data Repository

```python
from pycaret.datasets import get_data
get_data('index')
```

```python
data = get_data('juice')
```

# 2. Initialize Setup

```python
from pycaret.classification import *
clf1 = setup(data, target = 'Purchase', session_id=123, log_experiment=True, experiment_name='juice-v1')
```

# 3. Compare Baseline

```python
best_model = compare_models()
```

```python
top3 = compare_models(whitelist=['dt','rf','lightgbm'], n_select=3)
```

# 4. Create Model

```python
lr = create_model('lr')
```

```python
dt = create_model('dt', fold = 5)
```

```python
rf = create_model('rf', fold = 5)
```

```python
models()
```

```python
models(type='ensemble').index.tolist()
```

```python
ensembled_models = compare_models(whitelist = models(type='ensemble').index.tolist(), fold = 3)
```

# 5. Tune Hyperparameters

```python
tuned_lr = tune_model(lr)
```

```python
tuned_rf = tune_model(rf)
```

# 6. Ensemble Model

```python
dt
```

```python
bagged_dt = ensemble_model(dt)
```

```python
bagged_dt
```

```python
boosted_dt = ensemble_model(dt, method = 'Boosting')
```

```python
boosted_dt
```

# 7. Blend Models

```python
blender = blend_models(estimator_list = [bagged_dt, boosted_dt], method = 'soft')
```

```python
blender
```

# 8. Stack Models

```python
stacker = stack_models(estimator_list = [boosted_dt,bagged_dt,tuned_rf], meta_model=rf)
```
